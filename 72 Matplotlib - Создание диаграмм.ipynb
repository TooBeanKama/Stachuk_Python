# 72.01 Введение в Matplotlib и базовые диаграммы

```python
import matplotlib.pyplot as plt
import pandas as pd

values = [0,1,2,3,4,5,6,7]
squares = [value ** 2 for value in values]
print(squares)

plt.plot(values, squares);
```

```python
plt.scatter(values, squares)
plt.show()
```

```python
values = list(range(0,100))
powers = [pow(value, 3) for value in values]
plt.scatter(values, powers)
plt.show()
```

```python
values = list(range(0,100))
powers = [pow(value, 3) for value in values]
plt.scatter(values, powers, c=powers, cmap=plt.cm.Blues, edgecolors='none', s=10)
plt.show()
```

```python
values = list(range(50))
squares = [value ** 2 for value in values]
cubes = [value ** 3 for value in values]

# создаем подграфики: один под другим
fig, (ax1, ax2) = plt.subplots(2,1, sharex=True) # общая ось Х
ax1.scatter(values, squares)
ax1.set_title('Square numbers')

ax2.plot(values, cubes)
ax2.set_title('Cube numbers')

plt.show()
```

```python
values = list(range(10))
squares = [value ** 2 for value in values]
cubes = [value ** 3 for value in values]

# создаем подграфики: рядом
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True) # общая ось Y
ax1.scatter(values, squares)
ax1.set_title('Square numbers')

ax2.plot(values, cubes)
ax2.set_title('Cube numbers')

plt.show()
```

# 72.05 Пример диаграммы heatmap

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Week 1 Temp': [6, 10, 12, 14, 17],
    'Week 2 Temp': [8, 12, 13, 15, 18],
    'Week 3 Temp': [9, 11, 14, 16, 20],
    'Week 4 Temp': [10, 15, 10, 17, 22],
}

df = pd.DataFrame(data)
df
```

```python
# изменим индекс на имя месяца
df.set_index('Month', inplace=True)
df
```

```python
sns.heatmap(df)
plt.title('Weekly temperature data')
plt.show()
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly temperature data')
plt.show()
```
